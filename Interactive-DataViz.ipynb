## Exploring the Gapminder Dataset with Plotly Express

About the Data:
[Data Source](https://www.gapminder.org/tools/#$state$time$value=2007;;&chart-type=bubbles)

### Loading the Data 

```python
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import pandas as pd
import numpy as np
```

```python
import plotly.express as px
from plotly.figure_factory import create_table
```

```python
gapminder = px.data.gapminder()
table = create_table(gapminder.head(10))
py.iplot(table)
```

### Visualizations with Custom Bar Charts

Taking a look at the Population Growth in Bangladesh (1970-2007) by using a bar plot. Each row of the Dataframe is represented by a rectangular mark.

```python
data_bangladesh = px.data.gapminder().query("country == 'Bangladesh'")
fig = px.bar(data_bangladesh, x='year', y='pop', height=400)
fig.show()
```

```python
fig = px.bar(data_bangladesh, x='year', y='pop', hover_data=['lifeExp', 'gdpPercap'],
             color='lifeExp', labels={'pop':'population of Bangladesh'}, height=400, title="Population Growth in Bangladesh (1970-2007)")
fig.show()
```

### Plotting Life Expectancy vs GDP per Capita


```python
gapminder2007 = gapminder.query("year == 2007")
px.scatter(gapminder2007, x="gdpPercap", y="lifeExp")
```

```python
px.scatter(gapminder2007, x="gdpPercap", y="lifeExp", color='continent')
```

### Customizing Interactive Bubble Charts


```python
px.scatter(gapminder2007, x="gdpPercap", y="lifeExp", color='continent',
          size="pop", size_max=60)
```

```python
px.scatter(gapminder2007, x="gdpPercap", y="lifeExp", color='continent',
          size="pop", size_max=60, hover_name="country")
```

### Creating Interactive Animations and Facet Plots 


```python
px.scatter(gapminder2007, x="gdpPercap", y="lifeExp", color='continent',
          size="pop", size_max=60, hover_name="country", facet_col="continent", log_x=True)
```

```python
px.scatter(gapminder, x="gdpPercap", y="lifeExp", color='continent',
          size="pop", size_max=60, hover_name="country", animation_frame="year",
          animation_group="country", log_x=True, range_x=[100, 100000], range_y=[25, 90],
           labels = dict(pop="Population", gdpPercap="GDP per Capita",
                         lifeExp="Life Expectancy"))
```

### Representing Geographic Data as Animated Maps


```python
px.choropleth(gapminder, locations="iso_alpha", color="lifeExp", hover_name="country",
             animation_frame="year", color_continuous_scale=px.colors.sequential.Plasma,
             projection="natural earth")
```

```python
px.choropleth(gapminder, locations="iso_alpha", color="lifeExp", hover_name="country",
             animation_frame="year", color_continuous_scale=px.colors.sequential.Plasma,
             projection="orthographic")
```
